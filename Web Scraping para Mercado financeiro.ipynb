## Buscando indicadores de valuation de cada empresa listada na B3

Criando função para buscar informações de cada simbolo selecionado


```python
import requests, re
import pandas as pd
from bs4 import BeautifulSoup

def get_info_acao(ticker, nome):
  page = requests.get(
      f"https://www.dadosdemercado.com.br/bolsa/acoes/{ticker}"
      )
  soup = BeautifulSoup(page.content, 'html.parser')
  body = soup.body

  cotacao = body.select('h2')[1].text.strip()
  ramo    = body.select('div.about-params > div')[4].select('span')[1].text.strip()
  ult_att = body.select('p')[0].text.strip()
  if "Atualizado em" not in ult_att:
    ult_att = body.select('p')[1].text.strip()

  # Indicadores de valuation
  grid    = body.select('div.grid')[0]
  LPA     = grid.select('div.ratio')[0].select('span')[1].text.strip()
  VPA     = grid.select('div.ratio')[1].select('span')[1].text.strip()
  PL      = grid.select('div.ratio')[2].select('span')[1].text.strip()
  PVP     = grid.select('div.ratio')[3].select('span')[1].text.strip()

  data = {
      "empresa": nome,
      "simbolo": ticker,
      "cotacao diaria": cotacao.split("R$")[1],
      "ramo": ramo,
      "LPA": LPA,
      "VPA": VPA,
      "PL": PL,
      "P/VP": PVP,
      "ultima atualizacao": ult_att.split("Atualizado em")[1]
  }
  info_df = pd.json_normalize(data)
  return info_df
```

Montando dataframe com os dados da carteira

```python
from datetime import datetime
import pandas as pd

acoes.head(50)
carteira = pd.DataFrame()
for index, acao in acoes.iterrows():
  try:
    dados = get_info_acao(acao['Ticker'], acao['Nome'])
    carteira = pd.concat([carteira, dados], ignore_index=True)
  except Exception as error:
    print(f"{acao['Ticker']}: {error}")

carteira.to_excel(f"carteira-{datetime.today().strftime('%Y-%m-%d')}.xlsx",
              sheet_name='acoes', index=False)

```
